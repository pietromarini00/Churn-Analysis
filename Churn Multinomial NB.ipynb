```python
prior = df.groupby('Churn').size().div(len(df)) 
prior
```

```python
n_samples=df.Churn.count()
l=[]
churn=0
for i in range(n_samples):
    if df['Churn'][i]==1:
        l+=[df['gender'][i]]
        churn+=1
male=0
female=0
for i in l:
    if i==1:
        male+=1
    else:
        female+=1
perc=np.array([male/churn,female/churn])
print(perc[0])
print(perc[1])
```

```python
data=df
data
```

```python
likelihood = {}
likelihood['gender'] = data.groupby(['Churn', 'gender']).size().div(len(data)).div(prior)
likelihood['Partner'] = data.groupby(['Churn', 'Partner']).size().div(len(data)).div(prior)
likelihood['PhoneService'] = data.groupby(['Churn', 'PhoneService']).size().div(len(data)).div(prior)
likelihood['InternetService'] = data.groupby(['Churn', 'InternetService']).size().div(len(data)).div(prior)
likelihood['OnlineSecurity'] = data.groupby(['Churn', 'OnlineSecurity']).size().div(len(data)).div(prior)
likelihood['Contract'] = data.groupby(['Churn', 'Contract']).size().div(len(data)).div(prior)
likelihood['PaymentMethod'] = data.groupby(['Churn', 'PaymentMethod']).size().div(len(data)).div(prior)
```

```python
likelihood
```

```python
from sklearn.preprocessing import LabelEncoder
encoded_data = data.apply(LabelEncoder().fit_transform)
```

```python
from sklearn.naive_bayes import MultinomialNB
import numpy as np
clf = MultinomialNB()
clf.fit(encoded_data.drop(['Churn'], axis=1), encoded_data['Churn'])

# {"Age":'<=30', "Income":"medium", "Student":'yes' , "Credit_Rating":'fair'}
# The data is encoded as [1,2,1,1]
X = np.array([1,2,1,1,1,1,1,1,1])
print (clf._joint_log_likelihood(X.reshape(1,-1)))
print ("Prediction of : ", clf.predict(X.reshape(1,-1)))
```
